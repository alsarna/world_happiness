```python
rng = df["Score"].max() - df["Score"].min()
level_assign = round((rng/3), 3)
not_happy = df["Score"].min() + level_assign
medium_happy = not_happy + level_assign
level = []
for i in df.Score:
    if i < not_happy:
        level.append("Not happy")
    elif i >= not_happy and i < medium_happy:
        level.append("Medium happy")
    else:
        level.append("Happy")
df["Happiness level"] = level
```

```python
df.columns = ["Rank", "Reg", "Scor", "GDP", "SocSupp", "LifeEx", "Free", "Gene", "Corrup", "HapLvl"]
corr_check = px.scatter_matrix(df, dimensions = ["Rank", "GDP", "LifeEx", "Free", "Corrup"], color = "HapLvl")
corr_check.show()
```
